

# The "Majority Illusion" in Social Networks

> Lerman, K., Yan, X., & Wu, X.-Z. (2016). The ‚Äúmajority illusion‚Äù in social networks. _PLOS ONE, 11_(2), e0147617.  https://doi.org/10.1371/journal.pone.0147617


# 1.- Resumen Te√≥rico

## 1.1.- Problema
El trabajo estudia c√≥mo la estructura de las redes sociales puede distorsionar las percepciones individuales sobre la prevalencia de ciertas conductas, opiniones o atributos. Bajo condiciones espec√≠ficas, un estado que es raro a nivel global puede estar  sobrerrepresentado en los vecindarios locales de muchos individuos, fen√≥meno conocido como ‚Äúespejismo de la mayor√≠a‚Äù. Esto lleva a que las personas sobreestimen sistem√°ticamente la frecuencia de ese estado, facilitando la propagaci√≥n de contagios sociales y alterando el entendimiento de las normas y conductas predominantes.

## 1.2.- M√©todos
Los autores presentan un modelo  que cuantifica el espejismo de la mayor√≠a y lo validan mediante simulaciones y an√°lisis en redes sint√©ticas y reales. Utilizan modelos como el configuration model y el modelo Erd≈ës-R√©nyi . Luego manipulan propiedades como la distribuci√≥n de grados y la desasortatividad, y modifican la correlaci√≥n entre grado y atributo mediante procedimientos de intercambio de atributos entre nodos. Tambi√©n aplican el modelo a diferentes redes reales de colaboraci√≥n cient√≠fica, interacci√≥n gen√©tica y redes sociales, evaluando emp√≠ricamente la fuerza del espejismo de la mayor√≠a en cada red.

### 1.2.1.- Redes y datos

- **Sint√©ticas.** Se generan redes *scale-free* con **configuration model** (**N** $‚âà$ 10.000, exponentes **Œ±** ‚àà{2.1, 2.4, 3.1}) y redes **Erd≈ës‚ÄìR√©nyi** con grado medio comparable.
- **Reales.** Se analizan seis grafos emp√≠ricos heterog√©neos (colaboraci√≥n HepTh, Reactome, Digg, Enron, Twitter y blogs pol√≠ticos) para cubrir distintos niveles de asortatividad y dispersi√≥n de grados.

### 1.2.2.- Control de variables estructurales y de atributo
- **Asortatividad por grado ($r_{kk}$).** Se modifica mediante *rewiring* que preserva la secuencia de grados.
- **Correlaci√≥n grado‚Äìatributo ($\rho_{kx}$).** Se ajusta intercambiando etiquetas de $x$ entre nodos, manteniendo fija la prevalencia global $p = P(x{=}1)$ y la topolog√≠a.
- **Prevalencia global.** Se trabaja con valores $p$ ‚àà [0.05, 0.20].

### 1.2.3.- Modelo te√≥rico
Se deriva la probabilidad de que un nodo de grado $k$ observe m√°s de una fracci√≥n $\phi$ de vecinos activos, usando:
1. La distribuci√≥n conjunta de grados en extremos de aristas
2. La probabilidad condicional $(P(x{=}1 \mid k))$.


### 1.2.4.- Procedimiento experimental
1. **Construcci√≥n.** Para cada red (sint√©tica o real) se fija $p$, $r_{kk}$ y $\rho_{kx}$, y se asignan etiquetas $x$ a los nodos.
2. **Medici√≥n local.** Para cada nodo se calcula la fracci√≥n de vecinos activos y se registra si supera $\phi$ (por defecto, $\phi=0.5$).
3. **Agregaci√≥n.** Se estima la fracci√≥n de nodos que perciben mayor√≠a y se compara con la predicci√≥n del modelo te√≥rico.
4. **Contagio por umbral (robustez).** Como prueba de robustez, se ilustra el impacto de estos par√°metros en un proceso din√°mico: un nodo se activa si m√°s de la mitad de sus vecinos est√°n activos ($\phi=0.5$).

### 1.2.5.- M√©tricas y validaci√≥n
- **Estructura:** Distribuci√≥n de grados, grado medio $\langle k \rangle$, asortatividad $r_{kk}$.
- **Atributo:** Prevalencia $p$ y correlaci√≥n $\rho_{kx}$.
- **Validaci√≥n:** Concordancia teor√≠a‚Äìmedici√≥n en redes sint√©ticas y reales; promedios sobre m√∫ltiples realizaciones para aislar efectos de topolog√≠a vs. asignaci√≥n de atributos.

## 1.3.- Resultados

El estudio encuentra que el espejismo de la mayor√≠a es m√°s intenso en redes con distribuci√≥n de grado heterog√©nea y estructura desasortativa, en las que los nodos con gran conectividad tienden a adoptar el atributo destacado. Esto provoca que la mayor√≠a de los nodos perciba dicho atributo como mayoritario en su entorno, aun cuando es minoritario globalmente. Los resultados num√©ricos y emp√≠ricos demuestran que, en determinadas condiciones, hasta un 70% de los nodos puede experimentar este espejismo en redes reales, y que el grado de desasortatividad y la correlaci√≥n atributo-grado amplifican significativamente el fen√≥meno.

## 1.3.- Conclusiones

La prevalencia local puede diferir dr√°sticamente de la global por razones estructurales, generando ilusiones que sesgan percepciones, decisiones y din√°micas de asociaci√≥n. Algunas topolog√≠as son especialmente vulnerables: ligeros cambios en r‚Çñ‚Çñ o en œÅ‚Çñ‚Çì pueden amplificar tendencias colectivas pese a que el atributo sea raro a nivel global. Entender la topolog√≠a ‚Äîy su interacci√≥n con c√≥mo se distribuye el atributo‚Äî es clave para explicar la r√°pida difusi√≥n de normas impopulares o la emergencia de movimientos, y para anticipar creencias y comportamientos colectivos en sistemas sociales complejos.

## 2.1.- **Explicaci√≥n y Evaluaci√≥n del "Configuration Model"**

### 2.1.1.- ¬øQue es?

El **‚Äúconfiguration model‚Äù** es un m√©todo generador de redes aleatorias que permite construir redes con una distribuci√≥n de grados espec√≠fica. El procedimiento consiste en:
- Asignar a cada nodo un n√∫mero de conexiones (o ‚Äúhalf-edges‚Äù) seg√∫n la secuencia de grados que se desee.
- Conectar pares de ‚Äúhalf-edges‚Äù al azar, formando las aristas de la red, hasta usar todos los "half-edges" o ya no sea posible formar nuevos enlaces.
- Este modelo permite generar redes heterog√©neas, por ejemplo aquellas con distribuci√≥n de tipo ‚Äúscale-free‚Äù o ley de potencias, que replican la presencia de nodos altamente conectados (‚Äúhubs‚Äù), t√≠picos en muchas redes reales.

En el paper, los autores ocupan el configuration model para crear redes sint√©ticas que imiten la distribuci√≥n de grados de las redes sociales reales. Posteriormente, sobre estas redes estudian el fen√≥meno del ‚Äúmajority illusion‚Äù, analizando c√≥mo la distribuci√≥n de grados y la desasortatividad afectan la percepci√≥n local.

### 2.1.2.- Evaluaci√≥n
El **"configuration model"** es una excelente elecci√≥n para este paper porque permite crear redes con una distribuci√≥n de grados espec√≠fica, lo que es fundamental para capturar la heterogeneidad observada en muchas redes reales y analizar a fondo los efectos de la estructura en el ‚Äúmajority illusion‚Äù. Este modelo mantiene la secuencia de grados y ofrece flexibilidad para controlar la presencia de hubs y la disasortavilidad, dos propiedades clave que los autores demuestran que intensifican el fen√≥meno estudiado.

Para replicar lo mas cercano a los parametros indicados, este se debio realizar en un script python directo, ya que el uso de procesador es intenso para la generacion del grafo y para realizar el rewire

**Parametros:**

### 2.2.1.- Redes Erd≈ës‚ÄìR√©nyi (ER)
A diferencia de scale-free, ER no presenta heterogeneidad marcada en el grado; por eso el efecto depende sobre todo de la asortatividad por grado $r_{kk}$ y de la correlaci√≥n grado‚Äìatributo $œÅ_{kx}$. El experimento se paraleliz√≥ ‚Äúuna curva por proceso‚Äù, cada curva es un $‚ü®k‚ü©$, $P(x=1)$,$r_{kk}$ pasando por todos los $œÅ_{kx}$)

**Parametros**

- N√∫mero de nodos (N) = 10000
- Grado medio objetivo ($‚ü®k‚ü©$,avg_degrees) = {5.2, 2.5}  ‚Üí p = ‚ü®k‚ü©/(N‚àí1)
- Nodos Activos ($p_x$) = [0.05, 0.10, 0.20]
- L√≠mite de iteraciones de rewire (M) = 8  
- Tolerancia en $r_{kk}$ (tol) = 0.015
- M√°x. swaps para alcanzar $œÅ_{kx}$ (MAX_SWAPS) = 50000
- Semilla reproducci√≥n global (seed) = 12021974
- Procesos en paralelo (processes) = min(18, n√∫cleos disponibles)
- Tolerancia $r_{kk}$ $r_{okk}$ (rkk_tol) = 0.015
- Limites $œÅ_{kx}$ [min, max, nr puntos] (rho_kxs) = [0, 0.6, 8]
- Objetivos:
  - $‚ü®k‚ü©$=5.2 y $‚ü®k‚ü©$=2.5
    - $r_{kk}$ ‚àà [-0.50, 0.00, +0.50]
    - $P_{x}$ ‚àà [0.05, 0.10, 0.20]

```
G_base(‚ü®k‚ü©) = ER(N, p = ‚ü®k‚ü©/(N-1))
para cada ‚ü®k‚ü© en {5.2, 2.5}:
  G0 = G_base(‚ü®k‚ü©)
  para cada rkk in {-0.5, 0.0, +0.5}:
    rewire G0 ‚Üí rkk (tol=0.015, factor=8)
    para cada œÅ en linspace(0, 0.6, 8):
      X  = asigna_atributo(G0, px1, objetivo œÅ)   # hasta 50k swaps
      œÄ¬Ω = fracci√≥n de nodos con mayor√≠a (>50%) de vecinos x=1
      guardar(‚ü®k‚ü©, px1, rkk, œÅ, œÄ¬Ω)
graficar_resultados()
```

### 2.3.1.- Implementaci√≥n
A diferencia de las redes sint√©ticas (ER y scale-free), en redes reales coexisten distribuciones de grado sesgadas, comunidades y r‚Çñ‚Çñ ‚Äúde f√°brica‚Äù; por eso la ilusi√≥n de mayor√≠a depende del r‚Çñ‚Çñ observado y de la correlaci√≥n grado‚Äìatributo œÅ‚Çñ‚Çì. Para este experimento paralelizamos ‚Äúuna curva por proceso‚Äù: cada curva es una tupla ‚ü®red, P(x=1)‚ü© analizando todos los œÅ‚Çñ‚Çì sobre la GCC y, en redes dirigidas, usando enlaces mutuos (A‚ÜîB) para desdirigir.

- **Par√°metros**
  - Nodos activos $P(x=1)$  = [0.05, 0.10, 0.20, 0.30]
  - œÅ‚Çñ‚Çì = linspace(0, 0.6, 10)
  - Ajuste de œÅ‚Çñ‚Çì: m√°x. 10 000 swaps por curva
  - Semilla global: 12021974
  - Procesamiento en paralelo: min(18, n√∫cleos disponibles)

```
for red in {Reactome, HepTh, Enron, Blogs, Twitter, Digg}:
  G = cargar_edge_list(red)
  if red es dirigida: G = solo_mutuos(G)  # A‚ÜîB
  G = GCC(G)  # componente conexa gigante
  rkk = assortatividad_por_grado(G)
  (k, vecinos) = precompute(G)  # grados y listas de vecinos

  for œÅ in linspace(0, 0.6, 10):
    for px in [0.05, 0.10, 0.20, 0.30]:
      X = asignar_atributos_con_objetivo_œÅ(k, px, œÅ, max_swaps=10000)
      œÄ¬Ω = fracci√≥n_de_nodos_con_mayor√≠a_activa(vecinos, X)  # >50% vecinos con x=1
      guardar_csv(timestamp, red, N, L, rkk, px, œÅ_obj, œÅ_fin, œÄ¬Ω, tiempos)
plot_curvas_por_red()  # una figura con subplots, leyenda por px
```
