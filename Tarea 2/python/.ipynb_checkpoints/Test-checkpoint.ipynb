```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

# Secuencia de grados con ley de potencias
def degree_sequence(N, alpha, min_k=1, max_k=100):
    seq = (np.random.pareto(alpha, N) + 1).astype(int)
    seq = np.clip(seq, min_k, max_k)
    # Asegura grado par
    if np.sum(seq) % 2 != 0:
        seq[0] += 1
    return seq

# Configuration model
def create_graph(N, alpha):
    deg_seq = degree_sequence(N, alpha)
    G = nx.configuration_model(deg_seq)
    G = nx.Graph(G) # elimina multi-edges
    G.remove_edges_from(nx.selfloop_edges(G))
    return G

# Simulaci√≥n para graficar (mock, no reproduce datos exactos de autores)
def simulate_majority_prob(alpha, rkk_list):
    curves = []
    for rkk in rkk_list:
        kx_corr = np.linspace(0.0, 0.55, 50)
        # Simula la forma cualitativa observada: mayor rkk y menor alpha
        p_majority = (kx_corr ** 1.2) * (1 + 2 * rkk) * (2.5 - alpha)
        p_majority = np.clip(p_majority, 0, 1)
        curves.append((kx_corr, p_majority))
    return curves

alphas = [2.1, 2.4, 3.1]
rkk_vals = [
    [-0.35, -0.25, -0.15, -0.05],
    [-0.20, -0.10,  0.00,  0.10, 0.20],
    [-0.15, -0.05, 0.00, 0.30]
]

fig, axs = plt.subplots(1, 3, figsize=(17, 5))
markers = ['o', '^', 's', 'x', 'D']
for idx, alpha in enumerate(alphas):
    curves = simulate_majority_prob(alpha, rkk_vals[idx])
    for i, (corr, pm) in enumerate(curves):
        marker = markers[i % len(markers)]
        axs[idx].plot(corr, pm, marker, label=f"$r_{{kk}}$={rkk_vals[idx][i]}")
    axs[idx].set_xlabel("k-x correlation")
    axs[idx].set_ylabel("Probability of majority, $P_{>1/2}$")
    axs[idx].set_title(f"({chr(97+idx)}) $\\alpha$={alpha}")
    axs[idx].set_xlim(0, 0.6)
    axs[idx].set_ylim(0, 1)
    axs[idx].legend()
plt.tight_layout()
plt.show()
```
